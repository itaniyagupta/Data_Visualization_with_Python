```python
import pandas as pd
import numpy as np
```

```python
objects = ['Mirror', 'Wheel', 'Lemon', 'Box', 'Bread', 'Towel', 'Laptop', 'Pen', 'Bottle', 'Biscuit']
shape =   ['circle', 'circle', 'circle', 'square', 'square', 'rectangle','rectangle','rectangle', 'square', 'rectangle']
colour = ['silver', 'silver', 'yellow', 'brown', 'white', 'white', 'black', 'blue', 'red', 'brown']
quantity = [1,4,4,3, 10,2,1,7,6, 20]
```

```python
object_list = list(zip(objects, shape, colour, quantity))
object_list
```

```python
object_data = pd.DataFrame(object_list, columns = ["object", 'shape', 'colour', 'quantity'])
```

```python
### MultiIndex Dataframes
array = objects_data.values.T.tolist()
```

```python
array
```

```python
array_index = array[1:3]
```

```python
array_index
```

```python
index = pd.MultiIndex.from_arrays(array_index, names = ['shape', 'colour'])
```

```python
multi_data = pd.DataFrame({'object' : array[0], 'quantity' : array[3]}, index = index)
```

```python
multi_data
```

```python
multi_data.groupby(level=['shape']).sum()
```

```python
multi_data.groupby(level=['colour']).sum()
```

```python
multi_data.groupby(level=['shape', 'colour']).sum()
```

```python
multi_data.groupby(level=['colour', 'shape']).sum()
```

```python
multi_data.groupby(level=['colour', 'shape'])\
          .get_group(('white', 'rectangle'))
```

```python
multi_data.groupby([pd.Grouper(level = 'shape'), 'object']).sum()
```

```python
multi_data.groupby([pd.Grouper(level = 0), 'object']).sum()
```

```python
for name, group in objects_data.groupby('colour'):
    print('Name:' , name)
    print('Group:' , group)
    print()
```

```python
multi_data.groupby(level=['colour', 'shape']).size()
```

```python
multi_data.groupby(level=['colour', 'shape']).describe()
```

```python
###agregation functions

multi_data.groupby(level=['shape']).aggregate(np.sum)
```

```python
multi_data.groupby(level=['shape', 'colour']).aggregate(np.sum)
```

```python
multi_data.groupby(level=['shape'], 
                   as_index = False).agg(np.sum)
```

```python
multi_data.groupby(level=['shape', 'colour'], 
                   as_index = False).agg(np.sum)
```

```python
multi_data.groupby(level=['shape'])\
          .aggregate({'quantity' : [np.sum, np.mean, np.min, np.max],
                      'object' : np.max
                     })
```

```python
###########Filtering
```

```python
perf_data = pd.read_csv('datasets/StudentsPerformance.csv')
```

```python
perf_data.head()
```

```python
gender_filter = perf_data["gender"].isin(["female"])
gender_filter.head(10)
```

```python
perf_data[gender_filter].head()
```

```python
perf_data[~gender_filter].head()
```

```python
condition = perf_data["gender"].isin(["female"]) & \
            (perf_data["reading score"] > perf_data["writing score"])

condition.head()
```

```python
perf_data[condition].head(10)
```

```python
course_filter = perf_data["test preparation course"].isin(["completed"])
```

```python
course_filter.head()
```

```python
perf_data[gender_filter & course_filter].head(10)
```

```python
education_levels = ["associate's degree", "bachelor's degree"]
```

```python
parent_filter = perf_data["parental level of education"]\
                .isin(education_levels)

perf_data[parent_filter].head(10)
```

```python
perf_data.filter(items = ['gender', 'test preparation course', 
                          'math score', 'reading score', 'writing score']).head(10)
```

```python
perf_data.filter(regex = 'score',
                axis = 1).head(10)
```

```python
perf_data.filter(like = '14', axis = 0).head(10)
```

```python
groupA = perf_data['race/ethnicity'] == 'group A'

perf_data[groupA].head(10)
```
