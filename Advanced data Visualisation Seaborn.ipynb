```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.subplots(figsize=(12,6))


sns.distplot(emp_data['Reason for absence'], bins= 29, kde= False, rug = True)
plt.show()
```

```python
##if dataset contain categorial data

zomato_data = pd.read_csv('datasets/zomato.csv', encoding='latin-1')
```

```python
zomato_data.head()
```

```python
zomato_data.describe()
```

```python
facet = sns.FacetGrid(data = zomato_data, col = "Price range")
plt.show()
```

```python
facet.map(plt.hist, "Aggregate rating")
plt.show()
```

```python
facet.map(plt.scatter, "Aggregate rating", "Votes", color = "0.3")
plt.show()
```

```python
facet = sns.FacetGrid(data = zomato_data, col = "Price range", hue = "Currency")
facet.map(plt.scatter, "Aggregate rating", "Votes", alpha = 0.5)

facet.add_legend()
plt.show()
```

```python
facet = sns.FacetGrid(data = zomato_data, row ="Has Online delivery", col = "Price range", hue = "Currency",
                     margin_titles= True)
facet.map(plt.scatter, "Aggregate rating", "Votes", alpha = 0.5,)  #fit_reg = False)

facet.add_legend()
plt.show()
```

```python
facet = sns.FacetGrid(data = zomato_data, row ="Has Online delivery", col = "Price range", hue = "Currency",
                     margin_titles= True, gridspec_kws = dict(width_ratios = [2,3,5,3]))
facet.map(plt.scatter, "Aggregate rating", "Votes", alpha = 0.5,)  #fit_reg = False)

facet.add_legend()
plt.show()
```

```python
facet = sns.FacetGrid(data = zomato_data,col = "Currency")
                    
facet.map(sns.barplot, "Aggregate rating","Average Cost for two")

plt.ylim(0,2000)
plt.show()
```

```python
facet = sns.FacetGrid(data = zomato_data,col = "Currency", col_wrap = 3)
                    
facet.map(sns.barplot, "Aggregate rating","Average Cost for two")

plt.ylim(0,2000)
plt.show()
```

```python
facet = sns.FacetGrid(data = zomato_data,col = "Currency", col_wrap = 3, aspect = 1.5)
                    
facet.map(sns.barplot, "Aggregate rating","Average Cost for two", color = '#826EDF')

plt.ylim(0,2000)
plt.show()
```

```python
facet = sns.FacetGrid(data = zomato_data,col = "Currency", col_wrap = 3, aspect = 1.5)
                    
facet.map(sns.barplot, "Aggregate rating","Average Cost for two", color = '#826EDF')

facet.fig.subplots_adjust(wspace = 0.5, hspace = 0.5)
facet.set_axis_labels("Rating", "Cost for Two")

plt.ylim(0,2000)
plt.show()
```

```python
poll_data = pd.read_csv('datasets/pollution_data.csv')
poll_data = poll_data[:1000]

poll_data.head()
```

```python
sns.set_palette("Set1")
fig, axis = plt.subplots(figsize = (10,5))

sns.distplot(poll_data["SO2 1st Max Hour"])
sns.distplot(poll_data["CO 1st Max Hour"])

plt.show()
```

```python
with sns.color_palette("Accent"):
        fig, axis = plt.subplots(figsize = (10,5))
        
        sns.distplot(poll_data["SO2 1st Max Hour"])
        sns.distplot(poll_data["CO 1st Max Hour"])
plt.show()
```

```python
sns.set()

current_palette = sns.color_palette()

sns.palplot(current_palette)

plt.show()
```

```python
hist = plt.subplots(figsize = (12,6))

with plt.rc_context(sns.plotting_context("notebook", font_scale = 1.5, rc = {"lines.linewidth" : 2})):
    sns.kdeplot(poll_data["NO2 AQI"])
    sns.kdeplot(poll_data["NO2 AQI"], bw= 0.3, label = "bw:0.3" )
    sns.kdeplot(poll_data["NO2 AQI"], bw= 10, label = "bw:10" )
    sns.kdeplot(poll_data["NO2 AQI"], bw= 14, label = "bw:14" )
    sns.kdeplot(poll_data["NO2 AQI"], bw= 38, label = "bw:38" )
    
plt.legend()
plt.show()
```
